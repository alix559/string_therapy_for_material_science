# Test solubility router end-to-end

Walks through:

1. Load `.env`
2. Hit the three placeholder viz services
3. Ensure schema + seed `routes/router_graph.json`
4. Call `/controller` in-process (JWT disabled)

**Prereqs (separate terminals):**

```bash
pixi run serve-viz          # ports 8001–8003
# Postgres + Apache AGE must match ROUTER_DATABASE_URL
```

## 1. Environment

````python
from __future__ import annotations

import json
import os
from pathlib import Path

import httpx
from dotenv import load_dotenv
from starlette.testclient import TestClient

from string_therapy import create_app, ensure_schema, seed_router_db

repo = Path.cwd()
if not (repo / '.env').exists() and (repo.parent / '.env').exists():
    repo = repo.parent

load_dotenv(repo / '.env', override=True)
graph_path = repo / 'routes' / 'router_graph.json'

print('repo:', repo)
print('graph:', graph_path, 'exists=', graph_path.is_file())
print('ROUTER_DATABASE_URL set:', bool(os.getenv('ROUTER_DATABASE_URL')))
print('JWT_AUTH_DISABLED:', os.getenv('JWT_AUTH_DISABLED'))
print('LISETTE_MODEL:', os.getenv('LISETTE_MODEL') or '(unset)')
print('LISETTE_API_KEY set:', bool((os.getenv('LISETTE_API_KEY') or '').strip()))
````

## 2. Placeholder viz services

Expect `pixi run serve-viz` already running.

````python
VIZ = [
    ('scatter', 'http://127.0.0.1:8001'),
    ('heatmap', 'http://127.0.0.1:8002'),
    ('timeseries', 'http://127.0.0.1:8003'),
]

with httpx.Client(timeout=5.0) as client:
    for name, base in VIZ:
        try:
            r = client.get(f'{base}/health')
            print(f'{name:12} GET /health -> {r.status_code} {r.text[:120]}')
        except httpx.HTTPError as e:
            print(f'{name:12} GET /health -> FAIL ({e})')
````

````python
payload = {'message': 'show NaCl solubility placeholder'}

with httpx.Client(timeout=10.0) as client:
    for name, base in VIZ:
        url = f'{base}/data-visualizations/{name}'
        try:
            r = client.post(url, json=payload)
            body = r.json()
            print(f'{name:12} POST -> {r.status_code} viz={body.get("visualization")} placeholder={body.get("placeholder")}')
        except Exception as e:
            print(f'{name:12} POST -> FAIL ({e})')
````

## 3. Inspect router graph

````python
graph = json.loads(graph_path.read_text(encoding='utf-8'))
print(f"nodes={len(graph['nodes'])} edges={len(graph['edges'])} acl={len(graph.get('acl', []))}")
for n in graph['nodes']:
    kind = n.get('node_type')
    extra = f" → {n.get('method', 'post').upper()} {n.get('url')}" if kind == 'endpoint' else ''
    print(f"  [{n['id']}] {kind}: {n.get('intent')}{extra}")
````

## 4. Schema + seed (Postgres + Apache AGE)

````python
if not os.getenv('ROUTER_DATABASE_URL'):
    raise RuntimeError('Set ROUTER_DATABASE_URL in .env')

ensure_schema()
print('schema ready')
seed_router_db(graph_path)
print('seeded', graph_path)
````

## 5. Controller in-process

Uses `TestClient(create_app())` — no separate `serve_controller` process needed.
With `JWT_AUTH_DISABLED=true`, no Bearer token is required.

````python
client = TestClient(create_app())

r = client.get('/health')
print('GET /health', r.status_code, r.json())
````

````python
body = {
    'message': 'Make a scatter plot of NaCl solubility vs temperature',
    'response_format': 'json',
}

r = client.post('/controller', json=body)
print('POST /controller', r.status_code)
try:
    data = r.json()
    print(json.dumps(data, indent=2)[:3000])
except Exception:
    print(r.text[:2000])
````

## 6. Try the other viz intents (optional)

Re-run with different messages so Lisette can pick heatmap or timeseries.

````python
messages = [
    'Show a heatmap of solubility across solvents and temperatures',
    'Plot solubility as a time series over 60 minutes',
]

for msg in messages:
    print('\n===', msg)
    r = client.post('/controller', json={'message': msg, 'response_format': 'json'})
    print('status:', r.status_code)
    try:
        print(json.dumps(r.json(), indent=2)[:1500])
    except Exception:
        print(r.text[:1000])
````

## Checklist

| Step | OK when… |
|------|----------|
| Viz health | each returns `{"status":"ok",...}` |
| Viz POST | each returns `placeholder: true` + plot data |
| Schema/seed | prints `schema ready` / `seeded` |

| `/controller` | `200` and routed response (needs Lisette key + live viz servers) |
